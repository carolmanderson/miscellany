## The problem:
    
Write a function that prints each number from 1 to 100 on a new line. 

For each multiple of 3, print "Fizz" instead of the number. 

For each multiple of 5, print "Buzz" instead of the number. 

For numbers which are multiples of both 3 and 5, print "FizzBuzz" instead of the number.

# Vanilla python version

```python
def fizz_buzz(n):
        out = ""
        if n % 3 == 0:
            out += "Fizz"
        if n % 5 == 0:
            out += "Buzz"
        if out:
            return(out) 
        else:
            return(str(n))
        
        
for i in range(1, 101):
    print(fizz_buzz(i))        
```

# TensorFlow version

```python
import tensorflow as tf
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior() 
```

```python
x = tf.placeholder(tf.int32, shape=(),name='x')
output_stem = tf.constant("")
divisible_by_3 = tf.equal(tf.mod(x, 3), 0)
divisible_by_5 = tf.equal(tf.mod(x, 5), 0)
output = tf.cond(divisible_by_3, lambda: tf.add(output_stem, tf.constant("Fizz")), lambda: output_stem)
output = tf.cond(divisible_by_5, lambda: tf.add(output, tf.constant("Buzz")), lambda: output)
final_output = tf.cond(tf.equal(output_stem, output), lambda: tf.as_string(x), lambda: output)

with tf.Session() as sess:
    for i in range(1, 101):
        print(sess.run(final_output, feed_dict= {x: i}).decode())
```

# Test for expected outputs

```python
def fizz_buzz_tf(n, sess):
    x = tf.placeholder(tf.int32, shape=(),name='x')
    output_stem = tf.constant("")
    divisible_by_3 = tf.equal(tf.mod(x, 3), 0)
    divisible_by_5 = tf.equal(tf.mod(x, 5), 0)
    output = tf.cond(divisible_by_3, lambda: tf.add(output_stem, tf.constant("Fizz")), lambda: output_stem)
    output = tf.cond(divisible_by_5, lambda: tf.add(output, tf.constant("Buzz")), lambda: output)
    final_output = tf.cond(tf.equal(output_stem, output), lambda: tf.as_string(x), lambda: output)
    return sess.run(final_output, feed_dict= {x: n}).decode()
    
with tf.Session() as sess:
    for i in range(1, 101):
        assert fizz_buzz_tf(i, sess) == fizz_buzz(i)   
```
